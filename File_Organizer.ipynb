```python
import os
import shutil
import logging

# ✅ Replace this with your actual folder path (e.g., Downloads)
SOURCE_FOLDER = r"C:\Users\bhavs\Downloads"

# Categories with extensions
FILE_TYPES = {
    'Images': ['.jpg', '.jpeg', '.png', '.gif'],
    'Documents': ['.pdf', '.docx', '.txt', '.xlsx'],
    'Videos': ['.mp4', '.mov', '.avi'],
    'Music': ['.mp3', '.wav'],
    'Archives': ['.zip', '.rar', '.7z'],
    'Scripts': ['.py', '.js', '.html'],
    'Others': []
}

# Setup logging to a file
logging.basicConfig(
    filename='file_organizer.log',
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

def move_file(src_path, folder_name):
    dest_folder = os.path.join(SOURCE_FOLDER, folder_name)
    os.makedirs(dest_folder, exist_ok=True)
    dest_path = os.path.join(dest_folder, os.path.basename(src_path))

    try:
        shutil.move(src_path, dest_path)
        logging.info(f"Moved: {src_path} → {dest_path}")
        print(f"Moved: {src_path} → {dest_path}")
    except Exception as e:
        logging.error(f"Error moving file {src_path}: {e}")
        print(f"❌ Error moving file {src_path}: {e}")

def organize_files():
    for filename in os.listdir(SOURCE_FOLDER):
        filepath = os.path.join(SOURCE_FOLDER, filename)
        if os.path.isfile(filepath):
            file_ext = os.path.splitext(filename)[1].lower()
            moved = False

            for folder_name, extensions in FILE_TYPES.items():
                if file_ext in extensions:
                    move_file(filepath, folder_name)
                    moved = True
                    break

            if not moved:
                move_file(filepath, 'Others')

organize_files()
print("✅ Done organizing files!")
```
